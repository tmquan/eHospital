```python
### 
df_hosobenhan = pd.read_csv('data/2.hosobenhan.csv',
                         sep=';', 
                         error_bad_lines=False, 
                         warn_bad_lines=False)

selected  = [
#     'hosobenhanid',
    'patientid',
    'chandoanravien_code',
    'chandoanravien',
#     'patientname',
#     'patientcode',
#     'gioitinhcode',
#     'nghenghiepcode',
    'noilamviec',
#     'hc_xacode', 
    'hc_huyencode',
    'hc_tinhcode',
#     'hc_quocgiacode',
#     'hc_xaname',
#     'hc_huyenname',
#     'hc_tinhname',
#     'hc_quocgianame',

]

df_hosobenhan = df_hosobenhan[selected]
# df_hosobenhan.dropna(thresh=len(df_hosobenhan) - len(df_hosobenhan)/3, 
#                   axis=1, 
#                   inplace=True)
display(df_hosobenhan.dtypes)


df_hosobenhan[df_hosobenhan.select_dtypes(['object']).columns] = df_hosobenhan.select_dtypes(['object']).apply(lambda x: x.astype('category'))
display(df_hosobenhan.dtypes)
display(df_hosobenhan['chandoanravien_code'].cat.codes)
# categorical to numerical
#df_hosobenhan['chandoanravien_code'] = df_hosobenhan['chandoanravien_code'].astype('category')
df_hosobenhan['chandoanravien_num'] = df_hosobenhan['chandoanravien_code'].cat.codes
display(df_hosobenhan.dtypes)

df_hosobenhan.dropna(inplace=True)

display(df_hosobenhan.columns)
display(len(df_hosobenhan.columns))
display(df_hosobenhan)
```
